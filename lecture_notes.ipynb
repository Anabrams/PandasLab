##### freqency analysis - value counts

```python
# Sample DataFrame
data = {'Category': ['Electronics', 'Clothing', 'Electronics', 'Books', 'Books', 'Clothing']}

cat_df = pd.DataFrame(data)

cat_df.value_counts()
```

```python
# Create the dataframe
sales_data={"Devices":['Laptop','iPhone','LED','LCD','Smart-Phone','Washing-Machine'],
           'Brand':['Lenovo','Apple','Samsung','Samsung','Samsung','Whirpool'],
           'Sales':[1000,2000,4000,2000,1000,4000],
           'Profit':[500,1000,1000,1500,1000,1500],
           'Pices left':[5000,4000,4000,5000,5000,1000]}

sales_df = pd.DataFrame(sales_data)
# get frequency count per brand
freq = sales_df['Brand'].value_counts()

freq_prd = freq[freq > 1].index[0]

print(freq)
print('item appears more than once:',freq_prd)
```

```python
data = {'Score' :[34, 57, 786, 23, 234, None, 45, 23]}
scores_df = pd.DataFrame(data)
score_count = scores_df.value_counts(dropna=False) # dropna, tells fucntion to eithr include or not incldue NaN, Null, NA values
print('Score', score_count)
```

##### freqency analysis - value counts

```python
# Sample DataFrame
data = {'Category': ['Electronics', 'Clothing', 'Electronics', 'Books', 'Books', 'Clothing']}

cat_df = pd.DataFrame(data)

cat_df.value_counts()
```

```python
# Create the dataframe
sales_data={"Devices":['Laptop','iPhone','LED','LCD','Smart-Phone','Washing-Machine'],
           'Brand':['Lenovo','Apple','Samsung','Samsung','Samsung','Whirpool'],
           'Sales':[1000,2000,4000,2000,1000,4000],
           'Profit':[500,1000,1000,1500,1000,1500],
           'Pices left':[5000,4000,4000,5000,5000,1000]}

sales_df = pd.DataFrame(sales_data)
# get frequency count per brand
freq = sales_df['Brand'].value_counts()

freq_prd = freq[freq > 1].index[0]

print(freq)
print('item appears more than once:',freq_prd)
```

```python
data = {'Score' :[34, 57, 786, 23, 234, None, 45, 23]}
scores_df = pd.DataFrame(data)
score_count = scores_df.value_counts(dropna=False) # dropna, tells fucntion to eithr include or not incldue NaN, Null, NA values
print('Score', score_count)
```

```python
url='https://raw.githubusercontent.com/bprasad26/lwd/master/data/tesla_stock_prices.csv'

turl='https://raw.githubusercontent.com/bprasad26/lwd/master/data/tesla_stock_prices.csv'
# using parse_dates to change type from string into date
tesla_df = pd.read_csv(url, parse_dates=['Date'])
# tesla_df.to_csv('data/tesla_stocks.csv')
tesla_df.head()
# tesla_df.info()
```

```python
# add new column to dateframe based of of dates
# Add new columns to dataframe based off of dates
tesla_df['Day'] = tesla_df['Date'].dt.day
tesla_df['Month'] = tesla_df['Date'].dt.month
tesla_df['Year'] = tesla_df['Date'].dt.year
tesla_df['day_name'] = tesla_df['Date'].dt.day_name()

tesla_df.head(10)
```

```python
#grab max and min:
most_recent = tesla_df['Date'].min()
least_recent = tesla_df['Date'].max()
print(most_recent)
least_recent
```

```python
monthly_sales = tesla_df.groupby(['Year','Month'])['Volume'].sum().reset_index()

monthly_sales.head(15)
```

##### Filter using loc[] or query[]

```python
fil_df_date = tesla_df.loc[(tesla_df['Date'] >= '2010-08-01')]

fil_df_date
```

```python
fil_df_date2 = tesla_df.loc[(tesla_df['Date'] >= '2010-08-01') & (tesla_df['Date'] <= '2010-12-31')]
fil_df_date2
```

```python
fil_df = tesla_df.query("Date >= '2010-08-01' and Date <= '2010-12-31'")
fil_df
```

### Adding Cols to Data Frame

```python
import pandas as pd
import numpy as np
```

```python
data = {'Name': ['Jane', 'Princi', 'James', 'Fadi'],
        'Height': [5.1, 6.2, 5.1, 5.2],
        'Qualification': ['Msc', 'MA', 'Msc', 'Msc'],
        'Score 1' : [56,86,77,45],
        'Score 2' : [50,96,60,30]}

df = pd.DataFrame(data)
print('Data Frame Before New Col:')
df.head()
```

```python
address = ['NYC', 'NJ', 'CA', 'PA']
df['Address'] = address
print('after new column:')
df.head()
```

##### Creating a new column using an OLD column

```python
print('Before column add:')
df.head()
```

```python
df['total score'] = df['Score 1'] + df['Score 2']
print('after add:')
df.head()
```

```python
# to get avg score
df['Avg_Score'] = df['total score'] / 2
df.head()
```

##### Importing Specified columns and rearanging

```python
data = {'Name': ['Jane', 'Princi', 'James', 'Fadi'],
        'Height': [5.1, 6.2, 5.1, 5.2],
        'Qualification': ['Msc', 'MA', 'Msc', 'Msc'],
        'Score 1' : [56,86,77,45],
        'Score 2' : [50,96,60,30]}

df2 = pd.DataFrame(data, columns=['Score 1', 'Score 2', 'Name', 'Qualification'])
df2.head()
```

```python
values = [['Rohan', 455], ['Elvish', 250], ['John', 495],
          ['Sai', 400], ['Eric', 350], ['Adam', 450]]

df3 = pd.DataFrame(values, columns=['Name', 'Univ_marks'])
df3.head()
```

```python
df3['Percentage'] = (df3['Univ_marks'] /500) *100
df3.head()
```

##### using .insert() method to add a column

```python
df3.insert(1, 'Age', [1000, 300, 45, 60, 14,13])
df3.head()
```

```python
#using .assign to create a new column
df3 = df3.assign(location=['MiddleEarth', 'Forgotten Lands', 'NYC', 'Florida', 'Maine', 'Tokyo'])
df3.head()
```
